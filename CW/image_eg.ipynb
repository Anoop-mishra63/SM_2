```
##1 (sigmoid)
def sigmoid(Z):
  s=1/(1+np.exp(-Z))
  return s
```

```
##2 initialize parameter
def int_value(nx):  ##for dimension
  w=np.zeros([nx,1])  ##if u wanna randomly initialize type np.rand.random
  b=0
  return w,b
```

```
##3 
def fwd(w,b,X,Y):
  m=5                 #####Ask nihal to send pic
  Z=np.dot(w,X.T)+b
  A=sigmoid(Z)
  cost=-np.sum(Y*np.log(A)+(1-Y)*np.log(1-A))/m

  dw=np.dot(X,(A-Y).T)/m
  db=np.sum(A-Y)/m 

  cost=np.squeeze(cost)## we use squeeze command for
  grades={"dw":dw,"db":db}
  return grades, cost
  
```

```
##4 no. of iterations
def opt(w,b,X,Y,num_itr,lr,print_cost=False):
  cost1=[]  ##for list
  for i in range(num_itr):
    grades,cost=fwd(w,b,X,Y)

    dw=grades["dw"]   #### $ sign ka kaam jo R me tha
    db=grades["db"]

    w=w-lr*dw
    b=b-lr*db

    if i%100==0:
      cost1.append(cost)

  par={"w":w,"b":b}
  grades={"dw":dw,"db":db}  ## backward propagation value
  return par,grades,cost1 

```

```
w,b,X,Y=np.random.rand(1,n_x),0,np.random.rand(n_x,m),np.array([[1,1,0,0,1]])
value=opt(w,b,X,Y,200,0.01)
value
```

```
##5
def pred(w,b,X_test):
  m=X_test.shape[1]
  y_pred=np.zeros([1,m])
  w=w.reshape(X_test.shape[0],1)
  Z=np.dot(w.T,X_test)+b
  A=sigmoid(Z)

  for i in range(A.shape[1]):
    if (A[0][i]<0.05):
      y_pred[0][i]=0
    else:
       y_pred[0][i]=1
    return y_pred  
```
